```python
import os
import openai
from getpass import getpass

openai.api_key = getpass("OpenAI API Key: ")
os.environ["OPENAI_API_KEY"] = openai.api_key
```

```python

response = openai.embeddings.create(
            input="Your text string goes here", model="text-embedding-3-small"
        )
print(response)
print(response.data[0])
print(len(response.data[0].embedding))
print(response.data[0].embedding)
print(response.usage)
```

```python


chat_completion = openai.chat.completions.create(
    messages=[
        {
            "role": "user",
            "content": "Say this is a test",
        }
    ],
    model="gpt-3.5-turbo",
)
print(chat_completion)

print(chat_completion.created)
print(chat_completion.model)
print(chat_completion.usage)
print(chat_completion.choices)
print(len(chat_completion.choices))
print(chat_completion.choices[0].message)

print(chat_completion.choices[0].message.role)
print(chat_completion.choices[0].message.content)

print(chat_completion.choices[0].index)
print(chat_completion.choices[0].finish_reason)
```
